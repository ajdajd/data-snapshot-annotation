```python
from dsa.utils import load_json
import pandas as pd
```

```python
preds = load_json("data/unhcr_batch4_backup.json")
preds
```

```python
results = []
for x in range(len(preds)):
    annot = preds[x]["annotations"][0]
    for y in annot["result"]:
        results.append(y)
```

```python
len(results)
```

```python
results[0]
```

```python
norm_areas = {}
for x in results:
    try:
        # if x["origin"] in ["prediction-changed", "manual"]:
        width = x["value"]["width"]
        height = x["value"]["height"]
        o_width = x["original_width"]
        o_height = x["original_height"]

        id = x["id"]
        na = (width * height) / (o_width * o_height)
        norm_areas[id] = na
    except:
        continue
```

```python
df = pd.DataFrame.from_dict(norm_areas, orient="index", columns=["norm_area"]).reset_index()

df
```
