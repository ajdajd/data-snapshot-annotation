```python
import glob
import pandas as pd
```

# Load files

```python
files = glob.glob("data/results_*.jsonl")
files
```

```python
df = pd.concat([pd.read_json(x, lines=True) for x in files]).reset_index(drop=True)
df
```

# Cost

```python
df = df.join(pd.json_normalize(df["cost"]))

total = df["total_cost_usd"].sum()
mean = df["total_cost_usd"].mean()
std = df["total_cost_usd"].std()

print(f"Total cost: ${total:.2f}")
print(f"Ave cost per snapshot: ${mean:.4f}")
print(f"Std dev: ${std:.4f}")
```

```python
df.groupby("source").agg(
    count=("source", "count"),
    total_cost_usd=("total_cost_usd", "sum"),
)
```
