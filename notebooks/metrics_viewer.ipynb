```python
from dsa.constants import ROOT
from dsa.utils import load_json
import pandas as pd
```

```python
res_dir = ROOT / "data/evaluation_output/"
res_files = list(res_dir.glob("report*.json"))
res_files
```

```python
thresholds = [
    "0.5",
    # "0.75",
]
classes = [
    "Figure",
    "Table",
]
metrics = [
    "precision",
    "recall",
    "mean_iou",
    "mean_area_precision",
    "mean_area_recall",
]
```

```python
cols = []
for r in res_files:
    row = {}
    row["filename"] = r.name

    res = load_json(r)
    for t in thresholds:
        for c in classes:
            for m in metrics:
                row[f"[{c}] {m} @ {t}"] = res["metrics"][t]["per_class"][c][m]

    cols.append(row)
```

```python
df = pd.DataFrame(cols).set_index("filename").transpose()
df
```
